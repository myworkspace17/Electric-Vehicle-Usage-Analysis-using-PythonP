# Visualizing And Analysing the Data

# 1. Model Year Vs Electric Range

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Model Year vs Electric Range
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_new, x='Model Year', y='Electric Range', alpha=0.7)
plt.title('Electric Range vs Model Year')
plt.xlabel('Model Year')
plt.ylabel('Electric Range (miles)')
plt.show()
```

# 2.Vehicle Distribution of Model Year

```python
# Plot distribution of Model Year
plt.figure(figsize=(10, 6))
sns.countplot(data=df_new, x='Model Year')
plt.title('Distribution of Model Year')
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

### 3 .Electric Vehicle Type by Model Year

```python
# Filter for top 10 most recent years
recent_years = df_new['Model Year'].value_counts().head(10).index

# Plot the number of electric vehicle types for top 10 most recent years
plt.figure(figsize=(12, 6))
sns.countplot(data=df_new[df_new['Model Year'].isin(recent_years)], x='Model Year', hue='Electric Vehicle Type', palette='Set1')

# Add labels and title
plt.title('Number of Electric Vehicle Types by Recent Years')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

# 4 Electric Range Vs Make

```python
# Get the top 10 makes by the number of vehicles
top_makes = df_new['Make'].value_counts().head(10).index

# Filter data for the top makes
df_top_makes = df_new[df_new['Make'].isin(top_makes)]

# Plot for only the top makes
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_top_makes, x='Make', y='Electric Range', palette='Set2')

# Add labels and title
plt.title('Electric Range by Top Vehicle Makes')
plt.xlabel('Make')
plt.ylabel('Electric Range (miles)')
plt.xticks(rotation=90)
plt.show()
```

```python

# Filter for Battery Electric Vehicles (BEVs)
df_bev = df_new[df_new['Electric Vehicle Type'].str.contains("BEV", case=False, na=False)]

# Count BEVs by model
model_bev_count = df_bev['Model'].value_counts()

# Get the model with the most BEVs
top_model = model_bev_count.idxmax()
top_model_count = model_bev_count.max()

# Print the result
print(f"Most BEVs are for the model '{top_model}' with {top_model_count} vehicles.")

# Plot the number of BEVs by model
model_bev_count.plot(kind='bar', color='lightcoral', figsize=(14, 8))
plt.title('Number of BEVs by Model')
plt.xlabel('Model')
plt.ylabel('Number of BEVs')
plt.xticks(rotation=90)
plt.show()
```

```python


# Filter for Plug-in Hybrid Electric Vehicles (PHEVs)
df_phev = df_new[df_new['Electric Vehicle Type'].str.contains("PHEV", case=False, na=False)]

# Count the number of PHEVs by model
model_phev_count = df_phev['Model'].value_counts()

# Get the model with the highest number of PHEVs
top_phev_model = model_phev_count.idxmax()
top_phev_count = model_phev_count.max()

# Print the result
print(f"The model with the most PHEVs is '{top_phev_model}' with {top_phev_count} vehicles.")

# Plot the distribution of the number of PHEVs per model using a bar plot
model_phev_count.plot(kind='bar', color='lightseagreen', figsize=(14, 8))

# Add labels and title
plt.title('Number of PHEVs by Model')
plt.xlabel('Model')
plt.ylabel('Number of PHEVs')
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
